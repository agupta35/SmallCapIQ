We can join the table from compustat to this data where you can find fundamental data of each of stock or we can run a similar query seperately to about clutter. but this is pretty much how you should be looking to pull and analyze the data

```
russellData = db.raw_sql("""select b.htick as Ticker, a.date as Date, a.prc as Price
from crsp.msf as a left join crsp.msfhdr as b
on a.permno=b.permno
where b.htick in %(tkr)s
and date>='11/01/2018'
order by ticker, date""", params=tkr)

#The date in above code can be set to any date you want the data from

russellData.head(200)
```
